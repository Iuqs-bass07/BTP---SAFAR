```python
import pandas as pd
```

```python
df=pd.read_csv("/content/trajectory.csv")
```

```python
df
```

```python
grouped = df.groupby('ID')
```

```python
dfs = {}
for group_id, group_data in grouped:
    dfs[group_id] = group_data
```

```python
for key, value in dfs.items():
    print(f"DataFrame for ID {key}:")
    print(value)
    print("\n")
```

```python
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
for key, value in dfs.items():
    if key <= 5:
        plt.figure()
        plt.plot(value['Position_X'], value['Position_Y'], marker='o', linestyle='-', label=f'ID {key}')
        plt.title(f'Trajectory for ID {key}')
        plt.xlabel('Position X')
        plt.ylabel('Position Y')
        plt.legend()
        plt.grid(True)
        plt.show()
```

```python
for key, value in dfs.items():
    if key <= 5:

        value['Time'] = pd.to_datetime(value['Time'], format='%H:%M:%S,%f')

        plt.figure()
        plt.plot(value['Time'], value['Position_X'], marker='o', linestyle='-', label=f'Position X (ID {key})')
        plt.plot(value['Time'], value['Position_Y'], marker='o', linestyle='-', label=f'Position Y (ID {key})')
        plt.title(f'Trajectory for ID {key}')
        plt.xlabel('Time')
        plt.ylabel('Position')
        plt.legend()
        plt.grid(True)
        plt.show()
```
